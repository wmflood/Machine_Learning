```python
#This code box adresses Task 1
aDict = {}
with open ('cities.txt') as f:
  for line in f:
    (key, val1, val2) = line.split("|")
    aDict[key] = [float(val1), float(val2)]
```

```python
#The following code boxes adress Task 2
import folium
```

```python
m = folium.Map(location= [37.0902, -95.7129], zoom_start=4, tiles="Stamen Terrain")
for item in aDict.values():
  folium.Marker(item).add_to(m)
  
```

```python
#the following code baxes adress Task 3
import geopy
from geopy import distance
```

```python
new = []
for line in aDict:
  point1 = tuple(aDict[line])
  for item in aDict:
    point2 = tuple(aDict[item])
    if (point1 != point2):
      part= line,item,str(distance.distance(point1, point2).km)
      if part[2] not in new:
        new.append(str(part))
        with open('cities_distance.txt', 'a') as f:     
          text = "|".join(new)
          f.write(text+'\n')

#print(new)
```
