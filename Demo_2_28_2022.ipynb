```python
import geopandas as gpd
```

```python
nyc_b = gpd.read_file(gpd.datasets.get_path("nybb"))
```

```python
nyc_b.plot()
```

```python
centroids = nyc_b.geometry.centroid
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(figsize=(8,8))

nyc_b.plot(ax=axes, color="white", edgecolor="black")
centroids.plot(ax=axes, color="red", marker="o", markersize=20)
```

```python
nyc_convexhull = nyc_b.geometry.convex_hull

fig, axes = plt.subplots(figsize=(8,8))

nyc_b.plot(ax=axes, color="white", edgecolor="black")
nyc_convexhull.plot(ax=axes, cmap = plt.get_cmap("Set2"), alpha=0.2)
```

```python
nyc_b.crs
```

```python
nyc_buffer = nyc_b.geometry.buffer(5000)

fig, axes = plt.subplots(figsize=(8,8))

nyc_b.plot(ax=axes, color="white", edgecolor="black")
nyc_buffer.plot(ax=axes, cmap = plt.get_cmap("Set2"), alpha=0.5)
```

```python
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
countries.head()
```

```python
countries_geometry = countries[["geometry","iso_a3"]]
countries_attribute = countries[["name","iso_a3"]]
```

```python
countries_geometry.head()
```

```python
countries_attribute.head()
```

```python
joined_frame = countries_geometry.merge(countries_attribute, on="iso_a3")
```

```python
gpd.datasets.available
```

```python
countries.head()
```

```python
europe = countries[countries["continent"] == "Europe"]
europe.plot()
```

```python
cities = gpd.read_file(gpd.datasets.get_path("naturalearth_cities"))
cities.plot()
```

```python
cities_in_europe = gpd.sjoin(cities, europe, op="within")
cities_in_europe.plot()
```

```python
fig, axes = plt.subplots(figsize=(10,10))

europe.plot(ax=axes,color="white", edgecolor="black")
cities_in_europe.plot(ax=axes, marker="o", color="red", markersize=8)
```

```python
countries.plot(figsize=(8,8), color="white", edgecolor="black")
```

```python
continent = countries.dissolve(by="continent")
```

```python
continent.plot(figsize=(8,8), color="white", edgecolor="black")
```

```python
SouthAmerica = countries[countries["continent"] == "South America"]
```

```python
SouthAmerica.plot()
```

```python
cities.plot()
```

```python
cities_in_SA = gpd.sjoin(cities,SouthAmerica, op="within")
cities_in_SA.plot()
```

```python
city_buffer = cities_in_SA.geometry.buffer(1)
```

```python
type(city_buffer)
```

```python
city_buffer_gdf = gpd.GeoDataFrame(geometry=city_buffer)
```

```python
difference_SA = gpd.overlay(SouthAmerica, city_buffer_gdf, how="difference")
```

```python
difference_SA.plot()
```
