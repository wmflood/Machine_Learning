```python
def class_reader (x):
  for item in x:
    if x[item] == 513:
      return "Good"
    elif x[item] == 514:
      return "Fine"
    else:
      return "Oops"
```

```python
import random
```

```python
x = random.randint(0,6)
def play_game (a):
  if a == 5:
    return "Winner!"
  elif a == 4:
    return "Close"
  elif a == 6:
    return "Close"
  else:
    return "Loser"
```

```python
def chance_it (chance):
  if chance == 1:
    return "Safe!"
  elif 2 <= chance >= 10:
    return "Start at square 1"
  else:
    return "Goodbye!"
```

```python
big = random.randint(0,11)
chance_it (big)
```
